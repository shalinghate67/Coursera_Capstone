```python
collision = pd.read_csv("Data.csv", skiprows = 1, header = None)

collision.columns = collision.columns.astype(str)
```

```python
collision.drop(['3', '4', '5', '6', '9', '11', '12', '21', '25', '26', '30', '32', '33', '35', '36'], axis = 1, inplace = True)
collision
```

```python
collision.rename(columns = {'0':'Severity', '1':'X', '2': 'Y', '7':'Status', '8': 'ADDRTYPE', '10':'Location', '13':'SeverityCode', '14':'SeverityDescript', '15':'CollisionType', '16':'PersonCount', '17':'PedestrianCount', '18':'PedCylCount', '19':'VehicleCount', '20':'Date', '22':'JunctionType', '23':'SDOT_COLCODE', '24':'SDOT_COLDESC', '27':'Weather', '28':'RoadConditions', '29':'LightConditions', '31':'SDOTCOLNUM', '34':'STCOLDESC', '37':'HitParkedCar'}, inplace = True)
collision
```

```python
df = collision.loc[:, ['Severity','LightConditions']]
df.head()
```

```python
# Check the correlation between the severity and the light conditions
df_time = df.groupby(['Severity'])['LightConditions'].value_counts().to_frame()
df_time
```

```python
df1 = collision.loc[:, ['SeverityDescript', 'CollisionType']]
df1.head()
```

```python
# Comparing the severity result to the event
df_crash = df1.groupby(['SeverityDescript'])['CollisionType'].value_counts().to_frame()
df_crash
```
